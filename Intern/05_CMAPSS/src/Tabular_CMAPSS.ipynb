```python
import pandas as pd

# Tải dữ liệu
train_df = pd.read_csv('E:/IAD/Intern/05_CMAPSS/CMAPSSData/train_FD004.txt', delim_whitespace=True, header=None)
test_df = pd.read_csv('E:/IAD/Intern/05_CMAPSS/CMAPSSData/test_FD004.txt', delim_whitespace=True, header=None)

# Đặt tên cột (theo định dạng của bộ dữ liệu)
columns = ['unit_number', 'time_in_cycles'] + [f'sensor_{i}' for i in range(1, 25)]
train_df.columns = columns
test_df.columns = columns

# Xem một số hàng đầu tiên của dữ liệu
train_df.head()
```

```python
# Tính toán RUL (Remaining Useful Life)
rul = train_df.groupby('unit_number')['time_in_cycles'].max().reset_index()
rul.columns = ['unit_number', 'max_cycle']
train_df = train_df.merge(rul, on='unit_number', how='left')
train_df['RUL'] = train_df['max_cycle'] - train_df['time_in_cycles']

# Tạo các đặc trưng từ dữ liệu
features = ['time_in_cycles'] + [f'sensor_{i}' for i in range(1, 25)]
X_train = train_df[features]
y_train = train_df['RUL']

X_test = test_df[features]
```
