Để đánh giá hiệu quả của các mô hình dự đoán với biến liên tục, có nhiều phương pháp và chỉ số đánh giá được sử dụng phổ biến. Dưới đây là một số phương pháp và chỉ số chính:

### 1. Root Mean Squared Error (RMSE)
- **Công thức**:
  \[
  RMSE = \sqrt{\frac{1}{n} \sum_{i=1}^{n} (y_i - \hat{y}_i)^2}
  \]
- **Ý nghĩa**: Đo lường độ lệch trung bình bình phương giữa giá trị thực và giá trị dự đoán. Giá trị RMSE nhỏ cho thấy mô hình có độ chính xác cao.

### 2. Mean Absolute Error (MAE)
- **Công thức**:
  \[
  MAE = \frac{1}{n} \sum_{i=1}^{n} |y_i - \hat{y}_i|
  \]
- **Ý nghĩa**: Đo lường độ lệch trung bình tuyệt đối giữa giá trị thực và giá trị dự đoán. MAE ít bị ảnh hưởng bởi các giá trị ngoại lệ so với RMSE.

### 3. Mean Absolute Percentage Error (MAPE)
- **Công thức**:
  \[
  MAPE = \frac{100\%}{n} \sum_{i=1}^{n} \left| \frac{y_i - \hat{y}_i}{y_i} \right|
  \]
- **Ý nghĩa**: Đo lường lỗi dự đoán dưới dạng phần trăm, dễ hiểu và so sánh giữa các mô hình khác nhau.

### 4. R-squared (R²)
- **Công thức**:
  \[
  R^2 = 1 - \frac{\sum_{i=1}^{n} (y_i - \hat{y}_i)^2}{\sum_{i=1}^{n} (y_i - \bar{y})^2}
  \]
- **Ý nghĩa**: Đo lường phần trăm biến thiên của biến phụ thuộc được giải thích bởi mô hình. Giá trị R² nằm trong khoảng từ 0 đến 1, giá trị gần 1 cho thấy mô hình giải thích tốt biến thiên của dữ liệu.

### 5. Adjusted R-squared
- **Công thức**:
  \[
  \bar{R}^2 = 1 - \left( \frac{1-R^2}{n-k-1} \right)
  \]
  Trong đó, \(n\) là số lượng quan sát và \(k\) là số biến độc lập trong mô hình.
- **Ý nghĩa**: Điều chỉnh giá trị R² để phản ánh số lượng biến độc lập trong mô hình. Dùng để so sánh các mô hình với số lượng biến khác nhau.

### 6. Mean Squared Error (MSE)
- **Công thức**:
  \[
  MSE = \frac{1}{n} \sum_{i=1}^{n} (y_i - \hat{y}_i)^2
  \]
- **Ý nghĩa**: Tương tự RMSE nhưng không lấy căn bậc hai. Giá trị MSE nhỏ cho thấy mô hình dự đoán chính xác.

### 7. Median Absolute Error (MedAE)
- **Công thức**:
  \[
  MedAE = \text{median}(|y_i - \hat{y}_i|)
  \]
- **Ý nghĩa**: Đo lường độ lệch trung bình tuyệt đối giữa giá trị thực và giá trị dự đoán bằng cách lấy giá trị trung vị. Ít bị ảnh hưởng bởi các giá trị ngoại lệ.

### 8. Symmetric Mean Absolute Percentage Error (sMAPE)
- **Công thức**:
  \[
  sMAPE = \frac{100\%}{n} \sum_{i=1}^{n} \frac{|y_i - \hat{y}_i|}{(|y_i| + |\hat{y}_i|)/2}
  \]
- **Ý nghĩa**: Tương tự MAPE nhưng điều chỉnh để tránh các giá trị rất lớn hoặc rất nhỏ, phù hợp hơn trong một số tình huống.

### 9. Coefficient of Determination (COD)
- **Công thức**:
  \[
  COD = 1 - \frac{\sum_{i=1}^{n} (y_i - \hat{y}_i)^2}{\sum_{i=1}^{n} (|y_i - \hat{y}_i| + |\hat{y}_i - \bar{y}|)^2}
  \]
- **Ý nghĩa**: Tương tự R² nhưng được điều chỉnh để phản ánh sự phân tán của dữ liệu.

### Lựa chọn phương pháp phù hợp
- **RMSE và MAE**: Dùng khi muốn đo lường lỗi dự đoán dưới dạng giá trị thực tế.
- **R² và Adjusted R²**: Dùng để đánh giá tổng quát về khả năng giải thích của mô hình.
- **MAPE và sMAPE**: Dùng khi cần biểu diễn lỗi dưới dạng phần trăm và so sánh giữa các mô hình khác nhau.
- **MedAE**: Dùng khi dữ liệu chứa nhiều giá trị ngoại lệ.

Việc lựa chọn chỉ số đánh giá phù hợp phụ thuộc vào đặc điểm của dữ liệu và yêu cầu cụ thể của bài toán dự đoán.
