## 1️⃣ **Introduction (Giới thiệu)**  
**🎯 Mục tiêu:** Trình bày bối cảnh, vấn đề và khoảng trống nghiên cứu.  

🔹 **Bối cảnh & Tầm quan trọng:**  
- Dự đoán **Remaining Useful Life (RUL)** là yếu tố quan trọng trong **bảo trì dự đoán** (*Predictive Maintenance - PdM*), giúp giảm downtime và tối ưu chi phí vận hành.  
- Các mô hình dự đoán RUL truyền thống (LSTM, CNN, XGBoost) chưa đảm bảo **tính giải thích** và thường yêu cầu **thu thập dữ liệu tập trung**, gây rủi ro về quyền riêng tư.  

🔹 **Khoảng trống nghiên cứu:**  
- **Transformer** có tiềm năng vượt trội so với LSTM nhưng chưa được nghiên cứu sâu trong RUL.  
- **Federated Learning (FL)** giúp bảo vệ dữ liệu nhưng chưa được khai thác tốt với Transformer trong PdM.  
- **XAI (Explainable AI)** rất cần thiết để làm rõ lý do dự đoán, nhưng chưa được ứng dụng nhiều với Transformer trong FL.  

🔹 **Mục tiêu bài báo:**  
- Đề xuất **FL-Transformer** để dự đoán RUL mà không cần chia sẻ dữ liệu thô giữa các client.  
- Kết hợp **XAI (SHAP, Attention Visualization, Grad-CAM...)** để giải thích quyết định của mô hình.  
- So sánh mô hình với **LSTM, CNN, XGBoost** trên bộ dữ liệu tiêu chuẩn **C-MAPSS**.  

🔹 **Đóng góp chính:**  
✅ Lần đầu tiên tích hợp **FL + Transformer + XAI** để dự đoán RUL.  
✅ Đề xuất **cách làm sáng tỏ quyết định của Transformer** bằng XAI.  
✅ Cung cấp phân tích **sâu rộng** về hiệu suất và tính giải thích của mô hình.

## 3️⃣ **Methodology (Phương pháp đề xuất)**  
**🎯 Mục tiêu:** Mô tả chi tiết mô hình đề xuất **FL-Transformer với XAI**.  

### 🚀 **3.1. Kiến trúc FL-Transformer cho RUL**  
🔹 **Transformer Encoder:**  
- Xử lý dữ liệu dạng chuỗi thời gian của cảm biến động cơ.  
- **Self-Attention** giúp mô hình tập trung vào các đặc trưng quan trọng.  

🔹 **Federated Learning:**  
- Dữ liệu RUL phân tán giữa nhiều **client (máy bay, động cơ, nhà máy)**.  
- Sử dụng **FedAvg hoặc FedProx** để cập nhật mô hình mà không chia sẻ dữ liệu.  

🔹 **Mô hình tổng thể:**  
- Client: Training cục bộ trên dữ liệu riêng.  
- Server: Tổng hợp trọng số từ nhiều client.  


### 🔍 **3.2. Giải thích mô hình bằng XAI**  
🔹 **SHAP (SHapley Additive Explanations):**  
- Đánh giá mức độ quan trọng của từng đặc trưng cảm biến đối với dự đoán.  

🔹 **Attention Visualization:**  
- Kiểm tra Transformer tập trung vào đặc trưng nào khi đưa ra quyết định.  

🔹 **Grad-CAM/LRP:**  
- Hiển thị khu vực quan trọng trong input time-series.  


### 🔧 **3.3. Chi tiết bộ dữ liệu & Training**  
🔹 **Dataset: C-MAPSS (Commercial Modular Aero-Propulsion System Simulation)**  
- 100+ động cơ phản lực với dữ liệu cảm biến theo thời gian.  

🔹 **Evaluation Metrics:**  
- RMSE, MAE, Score Function để đánh giá hiệu suất.  

🔹 **Baseline Models:**  
- So sánh với LSTM, CNN, XGBoost để chứng minh hiệu quả.
